```python
df["avg_delinquency"] = (
    df[pay_history_cols]
    .clip(lower=0)
    .mean(axis=1)
)
```

```python
df["delinquency_trend"] = (
    df["PAY_0"].clip(lower=0)
    -
    df["PAY_6"].clip(lower=0)
)
```

```python
import pandas as pd
import numpy as np

from sklearn.feature_selection import f_classif, chi2, mutual_info_classif
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
target = "default.payment.next.month"

candidate_features = [
    "LIMIT_BAL",
    "AGE",
    "SEX",
    "EDUCATION",
    "MARRIAGE",

    "avg_bill",
    "avg_payment",
    "current_utilisation",
    "avg_utilisation",
    "payment_ratio",
    "has_bill",

    "delinquency_score",
    "max_delinquency",
    "number_of_times_delayed",
    "recent_delinquency"
]
```

```python
fs_df = final_df.copy()

X = fs_df[candidate_features]
y = fs_df[target]
```

```python
numeric_features = [
    "LIMIT_BAL",
    "AGE",
    "avg_bill",
    "avg_payment",
    "current_utilisation",
    "avg_utilisation",
    "payment_ratio",
    "delinquency_score",
    "max_delinquency",
    "number_of_times_delayed",
    "recent_delinquency"
]

f_scores, p_values = f_classif(X[numeric_features], y)

anova_results = pd.DataFrame({
    "feature": numeric_features,
    "f_score": f_scores,
    "p_value": p_values
}).sort_values(by="f_score", ascending=False)

anova_results
```

```python
categorical_features = [
    "SEX",
    "EDUCATION",
    "MARRIAGE",
    "has_bill"
]

X_cat = X[categorical_features].copy()

chi_scores, chi_p_values = chi2(X_cat, y)

chi_results = pd.DataFrame({
    "feature": categorical_features,
    "chi2_score": chi_scores,
    "p_value": chi_p_values
}).sort_values(by="chi2_score", ascending=False)

chi_results
```

```python
X_scaled = X.copy()

scaler = MinMaxScaler()
X_scaled[candidate_features] = scaler.fit_transform(X_scaled[candidate_features])

mi_scores = mutual_info_classif(
    X_scaled,
    y,
    random_state=42
)

mi_results = pd.DataFrame({
    "feature": candidate_features,
    "mutual_information": mi_scores
}).sort_values(by="mutual_information", ascending=False)

mi_results
```

```python
X_vif = X[numeric_features].copy()

X_vif = X_vif.replace([np.inf, -np.inf], np.nan)
X_vif = X_vif.fillna(X_vif.median())

vif_results = pd.DataFrame()
vif_results["feature"] = X_vif.columns

vif_results["VIF"] = [
    variance_inflation_factor(X_vif.values, i)
    for i in range(X_vif.shape[1])
]

vif_results.sort_values(by="VIF", ascending=False)
```

```python
rf_selector_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight="balanced"
)

rfe = RFE(
    estimator=rf_selector_model,
    n_features_to_select=8
)

rfe.fit(X, y)

rfe_results = pd.DataFrame({
    "feature": candidate_features,
    "selected": rfe.support_,
    "ranking": rfe.ranking_
}).sort_values(by="ranking")

rfe_results
```

```python
selection_summary = pd.DataFrame({
    "feature": candidate_features
})

selection_summary = selection_summary.merge(
    mi_results,
    on="feature",
    how="left"
)

selection_summary = selection_summary.merge(
    rfe_results[["feature", "selected", "ranking"]],
    on="feature",
    how="left"
)

selection_summary = selection_summary.merge(
    anova_results[["feature", "p_value"]],
    on="feature",
    how="left"
)

selection_summary = selection_summary.merge(
    chi_results[["feature", "p_value"]],
    on="feature",
    how="left",
    suffixes=("_anova", "_chi")
)

selection_summary.sort_values(
    by=["selected", "mutual_information"],
    ascending=[False, False]
)
```

```python
pd_results = X_test_lean.copy()

pd_results["actual_default"] = y_test_lean.values
pd_results["predicted_default"] = y_pred_final
pd_results["predicted_pd"] = y_pred_proba_final
```

```python
def assign_risk_segment(pd):
    if pd < 0.10:
        return "Low Risk"
    elif pd < 0.30:
        return "Medium Risk"
    else:
        return "High Risk"
```

```python
pd_results["risk_segment"] = pd_results["predicted_pd"].apply(assign_risk_segment)
```

```python
pd_results["risk_segment"].value_counts()
```

```python
def cli_decision(row):
    if row["predicted_pd"] < 0.10 and row["current_utilisation"] < 0.50:
        return "Increase"
    elif row["predicted_pd"] >= 0.30:
        return "Decrease"
    else:
        return "Maintain"
```

```python
pd_results["cli_decision"] = pd_results.apply(cli_decision, axis=1)
```

```python
pd_results["cli_decision"].value_counts()
```

```python
def limit_change_pct(row):
    if row["cli_decision"] == "Increase":
        return 0.20
    elif row["cli_decision"] == "Decrease":
        return -0.20
    else:
        return 0.0
```

```python
pd_results["limit_change_pct"] = pd_results.apply(limit_change_pct, axis=1)

pd_results["recommended_limit"] = (
    pd_results["LIMIT_BAL"] *
    (1 + pd_results["limit_change_pct"])
)
```

```python
MAX_INCREASE = 0.25
MAX_DECREASE = 0.20

MIN_LIMIT = 10000
MAX_LIMIT = 1000000
```

```python
pd_results["upper_guardrail"] = pd_results["LIMIT_BAL"] * (1 + MAX_INCREASE)
pd_results["lower_guardrail"] = pd_results["LIMIT_BAL"] * (1 - MAX_DECREASE)
```

```python
pd_results["guardrail_limit"] = pd_results["recommended_limit"].clip(
    lower=pd_results["lower_guardrail"],
    upper=pd_results["upper_guardrail"]
)

pd_results["final_recommended_limit"] = pd_results["guardrail_limit"].clip(
    lower=MIN_LIMIT,
    upper=MAX_LIMIT
)
```

```python
import shap
import pandas as pd

explainer = shap.TreeExplainer(best_xgb)

shap_values = explainer.shap_values(X_test_lean)

shap.summary_plot(
    shap_values,
    X_test_lean
)

shap.summary_plot(
    shap_values,
    X_test_lean,
    plot_type="bar"
)

shap_df = pd.DataFrame(
    shap_values,
    columns=X_test_lean.columns,
    index=X_test_lean.index
)

def get_top_reasons(row, top_n=3):
    return (
        row.abs()
        .sort_values(ascending=False)
        .head(top_n)
        .index
        .tolist()
    )

pd_results["reason_codes"] = (
    shap_df
    .apply(get_top_reasons, axis=1)
    .values
)

print(
    pd_results[
        [
            "predicted_pd",
            "risk_segment",
            "cli_decision",
            "final_recommended_limit",
            "reason_codes"
        ]
    ].head()
)
```
