```python
import pandas as pd
import numpy as np

from sklearn.feature_selection import f_classif, chi2, mutual_info_classif
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
target = "default.payment.next.month"

candidate_features = [
    "LIMIT_BAL",
    "AGE",
    "SEX",
    "EDUCATION",
    "MARRIAGE",

    "avg_bill",
    "avg_payment",
    "current_utilisation",
    "avg_utilisation",
    "payment_ratio",
    "has_bill",

    "delinquency_score",
    "max_delinquency",
    "number_of_times_delayed",
    "recent_delinquency"
]
```

```python
fs_df = final_df.copy()

X = fs_df[candidate_features]
y = fs_df[target]
```

```python
numeric_features = [
    "LIMIT_BAL",
    "AGE",
    "avg_bill",
    "avg_payment",
    "current_utilisation",
    "avg_utilisation",
    "payment_ratio",
    "delinquency_score",
    "max_delinquency",
    "number_of_times_delayed",
    "recent_delinquency"
]

f_scores, p_values = f_classif(X[numeric_features], y)

anova_results = pd.DataFrame({
    "feature": numeric_features,
    "f_score": f_scores,
    "p_value": p_values
}).sort_values(by="f_score", ascending=False)

anova_results
```

```python
categorical_features = [
    "SEX",
    "EDUCATION",
    "MARRIAGE",
    "has_bill"
]

X_cat = X[categorical_features].copy()

chi_scores, chi_p_values = chi2(X_cat, y)

chi_results = pd.DataFrame({
    "feature": categorical_features,
    "chi2_score": chi_scores,
    "p_value": chi_p_values
}).sort_values(by="chi2_score", ascending=False)

chi_results
```

```python
X_scaled = X.copy()

scaler = MinMaxScaler()
X_scaled[candidate_features] = scaler.fit_transform(X_scaled[candidate_features])

mi_scores = mutual_info_classif(
    X_scaled,
    y,
    random_state=42
)

mi_results = pd.DataFrame({
    "feature": candidate_features,
    "mutual_information": mi_scores
}).sort_values(by="mutual_information", ascending=False)

mi_results
```

```python
X_vif = X[numeric_features].copy()

X_vif = X_vif.replace([np.inf, -np.inf], np.nan)
X_vif = X_vif.fillna(X_vif.median())

vif_results = pd.DataFrame()
vif_results["feature"] = X_vif.columns

vif_results["VIF"] = [
    variance_inflation_factor(X_vif.values, i)
    for i in range(X_vif.shape[1])
]

vif_results.sort_values(by="VIF", ascending=False)
```

```python
rf_selector_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight="balanced"
)

rfe = RFE(
    estimator=rf_selector_model,
    n_features_to_select=8
)

rfe.fit(X, y)

rfe_results = pd.DataFrame({
    "feature": candidate_features,
    "selected": rfe.support_,
    "ranking": rfe.ranking_
}).sort_values(by="ranking")

rfe_results
```

```python
selection_summary = pd.DataFrame({
    "feature": candidate_features
})

selection_summary = selection_summary.merge(
    mi_results,
    on="feature",
    how="left"
)

selection_summary = selection_summary.merge(
    rfe_results[["feature", "selected", "ranking"]],
    on="feature",
    how="left"
)

selection_summary = selection_summary.merge(
    anova_results[["feature", "p_value"]],
    on="feature",
    how="left"
)

selection_summary = selection_summary.merge(
    chi_results[["feature", "p_value"]],
    on="feature",
    how="left",
    suffixes=("_anova", "_chi")
)

selection_summary.sort_values(
    by=["selected", "mutual_information"],
    ascending=[False, False]
)
```

```python
lean_features = [
    "delinquency_score",
    "recent_delinquency",
    "payment_ratio",
    "avg_payment",
    "current_utilisation",
    "max_utilisation",
    "LIMIT_BAL"
]
```

```python
X_lean = pd_model_df[lean_features]

y = pd_model_df["default.payment.next.month"]
```

```python
from sklearn.model_selection import train_test_split

X_train_lean, X_test_lean, y_train_lean, y_test_lean = train_test_split(
    X_lean,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)
import pandas as pd
import numpy as np
```

```python
scale_pos_weight = (
    y_train_lean.value_counts()[0]
    /
    y_train_lean.value_counts()[1]
)

scale_pos_weight
```

```python
xgb_weighted = XGBClassifier(
    objective="binary:logistic",
    eval_metric="auc",
    random_state=42,
    scale_pos_weight=scale_pos_weight,
    n_estimators=300,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8
)

xgb_weighted.fit(X_train_lean, y_train_lean)
```

```python
y_pred_weighted = xgb_weighted.predict(X_test_lean)
y_pred_proba_weighted = xgb_weighted.predict_proba(X_test_lean)[:, 1]
```

```python
weighted_metrics = {
    "Model": "Weighted XGBoost",
    "Accuracy": accuracy_score(y_test_lean, y_pred_weighted),
    "Precision": precision_score(y_test_lean, y_pred_weighted),
    "Recall": recall_score(y_test_lean, y_pred_weighted),
    "F1": f1_score(y_test_lean, y_pred_weighted),
    "ROC_AUC": roc_auc_score(y_test_lean, y_pred_proba_weighted)
}

weighted_metrics
```

```python
print(classification_report(y_test_lean, y_pred_weighted))
confusion_matrix(y_test_lean, y_pred_weighted)
```

```python
thresholds = [0.5, 0.4, 0.3, 0.25, 0.2, 0.15]

weighted_threshold_results = []

for threshold in thresholds:
    y_pred_thresh = (y_pred_proba_weighted >= threshold).astype(int)

    weighted_threshold_results.append({
        "threshold": threshold,
        "accuracy": accuracy_score(y_test_lean, y_pred_thresh),
        "precision": precision_score(y_test_lean, y_pred_thresh),
        "recall": recall_score(y_test_lean, y_pred_thresh),
        "f1_score": f1_score(y_test_lean, y_pred_thresh)
    })

weighted_threshold_df = pd.DataFrame(weighted_threshold_results)

weighted_threshold_df
```

```python
weighted_threshold_df.sort_values(
    by="f1_score",
    ascending=False
)
```

```python
best_threshold = weighted_threshold_df.sort_values(
    by="f1_score",
    ascending=False
).iloc[0]["threshold"]

best_xgb = xgb_weighted

y_pred_proba_final = y_pred_proba_weighted

y_pred_final = (
    y_pred_proba_final >= best_threshold
).astype(int)

best_threshold
```

```python
final_metrics = {
    "Model": "Final Weighted XGBoost",
    "Threshold": best_threshold,
    "Accuracy": accuracy_score(y_test_lean, y_pred_final),
    "Precision": precision_score(y_test_lean, y_pred_final),
    "Recall": recall_score(y_test_lean, y_pred_final),
    "F1": f1_score(y_test_lean, y_pred_final),
    "ROC_AUC": roc_auc_score(y_test_lean, y_pred_proba_final)
}

final_metrics
```

```python
feature_importance_final = pd.DataFrame({
    "feature": X_train_lean.columns,
    "importance": best_xgb.feature_importances_
}).sort_values(
    by="importance",
    ascending=False
)

feature_importance_final
```

```python
pd_results = X_test_lean.copy()

pd_results["actual_default"] = y_test_lean.values
pd_results["predicted_default"] = y_pred_final
pd_results["predicted_pd"] = y_pred_proba_final
```

```python
def assign_risk_segment(pd):
    if pd < 0.10:
        return "Low Risk"
    elif pd < 0.30:
        return "Medium Risk"
    else:
        return "High Risk"
```

```python
pd_results["risk_segment"] = pd_results["predicted_pd"].apply(assign_risk_segment)

pd_results["risk_segment"].value_counts()
```

```python
pd_results.groupby("risk_segment")["actual_default"].mean()
```

```python
def cli_decision(row):
    if row["predicted_pd"] < 0.10 and row["current_utilisation"] < 0.50:
        return "Increase"
    elif row["predicted_pd"] >= 0.30:
        return "Decrease"
    else:
        return "Maintain"
```

```python
pd_results["cli_decision"] = pd_results.apply(cli_decision, axis=1)

pd_results["cli_decision"].value_counts()
```

```python
def limit_change_pct(row):
    if row["cli_decision"] == "Increase":
        return 0.20
    elif row["cli_decision"] == "Decrease":
        return -0.20
    else:
        return 0.0
```

```python
pd_results["limit_change_pct"] = pd_results.apply(limit_change_pct, axis=1)

pd_results["recommended_limit"] = (
    pd_results["LIMIT_BAL"] *
    (1 + pd_results["limit_change_pct"])
)
```

```python
MAX_INCREASE = 0.25
MAX_DECREASE = 0.20

MIN_LIMIT = 10000
MAX_LIMIT = 1000000
```

```python
pd_results["upper_guardrail"] = pd_results["LIMIT_BAL"] * (1 + MAX_INCREASE)
pd_results["lower_guardrail"] = pd_results["LIMIT_BAL"] * (1 - MAX_DECREASE)

pd_results["guardrail_limit"] = pd_results["recommended_limit"].clip(
    lower=pd_results["lower_guardrail"],
    upper=pd_results["upper_guardrail"]
)

pd_results["final_recommended_limit"] = pd_results["guardrail_limit"].clip(
    lower=MIN_LIMIT,
    upper=MAX_LIMIT
)
```

```python
import shap

explainer = shap.TreeExplainer(best_xgb)

shap_values = explainer.shap_values(X_test_lean)

shap.summary_plot(
    shap_values,
    X_test_lean
)

shap.summary_plot(
    shap_values,
    X_test_lean,
    plot_type="bar"
)

shap_df = pd.DataFrame(
    shap_values,
    columns=X_test_lean.columns,
    index=X_test_lean.index
)

def get_top_reasons(row, top_n=3):
    return (
        row.abs()
        .sort_values(ascending=False)
        .head(top_n)
        .index
        .tolist()
    )

pd_results["reason_codes"] = (
    shap_df
    .apply(get_top_reasons, axis=1)
    .values
)

pd_results[
    [
        "predicted_pd",
        "risk_segment",
        "cli_decision",
        "final_recommended_limit",
        "reason_codes"
    ]
].head()
```
